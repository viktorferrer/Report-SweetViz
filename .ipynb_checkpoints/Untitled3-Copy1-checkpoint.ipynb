```python
# importing necessaries libs to explore tha data and build the ML model

# data analysis manipulation
import pandas as pd
import numpy as np

# data viz
from matplotlib import pyplot as plt
import seaborn as sns

# machine learning
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import ensemble, svm

# get better visualisation on Jupyter
%matplotlib inline
```

```python
# read datasets
train = pd.read_csv('/home/viktor/Documentos/DED/train.csv')
test = pd.read_csv('/home/viktor/Documentos/DED/test.csv')
```

```python
# 1 means that the passenger was alone
train['Family_Size']=train['SibSp']+train['Parch']+1
test['Family_Size']=test['SibSp']+test['Parch']+1

# converting data types to improve data manipulation
def convert_s(sex):
    if sex == 'female':
        return 0
    else:
        return 1

def convert_fam(fam):
    if fam == 1:
        return 1
    else:
        return 2

# adding new features to dataset
train['Gender'] = train['Sex'].map(convert_s)
test['Gender'] = test['Sex'].map(convert_s)

train['Family'] = train['Family_Size'].map(convert_fam)
test['Family'] = test['Family_Size'].map(convert_fam)

# replacing NaN by interpolated values


# map sex to a numeric type
# df.sex = df.sex.map({'male': 1, 'female': 0})

# Populate Age_Fill
train['Age_Filled'] = train['Age'].groupby([train['Gender']]).apply(lambda x: x.fillna(x.mean()))

# series with filled ages
groupby_result = train.Age_Fill[train.Age.isnull()]

display(train[train.Age.isnull()].head())

display(train[train.Fare.isnull()].head())
```
